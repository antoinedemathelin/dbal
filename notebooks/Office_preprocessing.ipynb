```python
import numpy as np
import tensorflow as tf
from PIL import Image
```

First download the Office dataset at https://people.eecs.berkeley.edu/~jhoffman/domainadapt/#datasets_code

```python
resnet50_notop = tf.keras.applications.ResNet50(include_top=False, pooling="avg")
```

```python
resnet50_notop.summary()
```

```python
import os

domain = "webcam"

# path to dataset
path_to_office = "../datasets/office/"
path = path_to_office + domain + "/images/"

features_list = []
labels_list = []
j = 0

for r, d, f in os.walk(path):
    for direct in d:
        print(direct)
        for r, d, f in os.walk(os.path.join(path , direct)):
            for file in f:
                path_to_image = os.path.join(r, file)
                image = Image.open(path_to_image)
                image = image.resize((224, 224), Image.ANTIALIAS)
                image = np.array(image)
                
                features = resnet50_notop.predict(image.reshape((1,) + image.shape))
                
                features_list.append(features)
                labels_list.append(direct)

                print(j, direct)
                j += 1
```

```python
features_array = np.concatenate(features_list, 0)
labels_array = np.array(labels_list)

np.save("../datasets/office_preprocessed/" + domain + "_X", features_array)
np.save("../datasets/office_preprocessed/" + domain + "_y", labels_array)
```
